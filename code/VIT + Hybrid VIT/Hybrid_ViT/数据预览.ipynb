```python
from PIL import Image
import os
from matplotlib import pyplot as plt
```

```python
fileroot = r'./data/Bitmojidata/trainimages/'
```

```python
fig, axes = plt.subplots(1, 4)
axes_list = []
for i in range(4):
    axes_list.append(axes[i])
for ind,file in enumerate(os.listdir(fileroot)):
    file_path = fileroot+file
    pic = Image.open(file_path)
    print(pic)
    axes_list[ind].imshow(pic)
    if ind == 3:
        break
```
